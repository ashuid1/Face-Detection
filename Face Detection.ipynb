```python
import cv2 as cv
import numpy as np
```

### Face detection

- Here we open the camera in real time then by fixing the coordinates or the size of the camera we can clik the images at real time
- to come out from this click on keyboard --> 'enter'

```python
classifier = cv.CascadeClassifier(r"C:\Users\user\Downloads\haarcascade_frontalface_default.xml")

cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img,1)
    
    
    try:
        faces = classifier.detectMultiScale(img, 1.1,8) 
        for(x,y,w,h) in faces:
            cv.rectangle(img, (x,y),(x+w,y+w),(0,180,0),5)
            
            
    except:
        pass
    
    cv.imshow('Frame', img)
    
    
    
    if(cv.waitKey(1) == 27):
        cam.release()
        break
```

```python
cv.destroyAllWindows()
```

### Croping the detected face

- After the clicking from the camera now Crop the image by palying with the index values of the generated image

```python
classifier = cv.CascadeClassifier(r"C:\Users\user\Downloads\haarcascade_frontalface_default.xml")

cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img,1)
    
    faces = classifier.detectMultiScale(img, 1.1,5) 
    
    for face in faces:
        if face[-1] == max(faces[:,-1]):
            print(face)
    if len(faces) >= 1 :
        for(x,y,w,h) in faces:
            cv.rectangle(img, (x,y),(x+w,y+w),(0,180,0),5)
            face = img[y:y+h, x:x+w]
     
    cv.imshow('Frame', img)
    cv.imshow('Face', face)
    
    
    
    if(cv.waitKey(1) == 27):
        cam.release()
        break
```

### Blur the detected face

- For blur the image just decrease the pixel value of the each index value of the image 
- adjust the index value as your reqired need.

```python
classifier = cv.CascadeClassifier(r"C:\Users\user\Downloads\haarcascade_frontalface_default.xml")

cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img,1)
    
    faces = classifier.detectMultiScale(img, 1.1,5) 
    
    for face in faces:
        if face[-1] == max(faces[:,-1]):
            print(face)
    if len(faces) >= 1 :
        for(x,y,w,h) in faces:
#            cv.rectangle(img, (x,y),(x+w,y+w),(0,180,0),5)
            face = img[y-50:y+h+50, x:x+w]
            face = cv.blur(face, (32,32))
        img[y-50:y+h+50,x:x+w] = face
    
     
    cv.imshow('Frame', img)
    cv.imshow('Face', face)
    
    
    
    if(cv.waitKey(1) == 27):
        cam.release()
        break
```

### RealTime Face Black   By Rectangle

- Many times we see in videos the face of the object is blacked 
- We can also do it by changing all the pixel values to the 0 in RGB
- That will help to get the black image face

```python
classifier = cv.CascadeClassifier(r"C:\Users\user\Downloads\haarcascade_frontalface_default.xml")

cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img,1)
    
    faces = classifier.detectMultiScale(img, 1.1,5) 
    
    for face in faces:
        if face[-1] == max(faces[:,-1]):
            break
            
    
    for(x,y,w,h) in faces:
        face = img[y:y+h,x:x+w]
        
    black = np.zeros((face.shape), dtype = int)
    
    img[y:y+h,x:x+w] = black
    
    cv.imshow('Frame', img)
    ##cv.imshow('Face', face)
    
    if(cv.waitKey(1) == 27):
        cam.release()
        break
```

### By Circle Black
- Done the same job but by using the circle
- to apply the circle find the middle value by the coordinates of ranctangle and then chhose the radius

```python
classifier = cv.CascadeClassifier(r"C:\Users\user\Downloads\haarcascade_frontalface_default.xml")

cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    img = cv.flip(img,1)
    
    faces = classifier.detectMultiScale(img, 1.1,5) 
    
    for face in faces:
        if face[-1] == max(faces[:,-1]):
            break
            
    
    x = face[0]
    y = face[1]
    w = face[2]
    h = face[3]
    
    circle_x = x + int(w/2)
    circle_y = y + int(h/2)
    
    cv.circle(img,(circle_x,circle_y), int(w/1.5), (0,180,0),-1)
    
    cv.imshow('Frame', img)
    
    if(cv.waitKey(1) == 27):
        cam.release()
        break
```

```python
cv.destroyAllWindows()
```

### Croping the face from the image
- In this task crop faces of the multiple man detected from the image
- just load the image path then by enter and delete choose the croped image that will be stored or not
- After croping the image store it into the local file/ folder

```python
import os
```

```python
img = cv.imread(r'C:\Users\user\OneDrive\Pictures\IMG_20230108_112132_690.jpg')

classifier = cv.CascadeClassifier(r"C:\Users\user\Downloads\haarcascade_frontalface_default.xml")
faces = classifier.detectMultiScale(img, 1.1,5)

def save(frame, folder_name):
    name_img = len(os.listdir(folder_name)) + 1
    name_img = folder_name + '/IMG_' + str(name_img) + '.png'
    cv.imwrite(name_img, frame)
    print(name_img , ' is exported')

for(x,y,w,h) in faces:
    face = img[y:y+h,x:x+w]
    cv.imshow('Face', face)
    
    if cv.waitKey(0) == 13:         ## Save Image 13 -> enter
        save(face,r"C:\Users\user\OneDrive\Desktop\New folder")
    elif cv.waitKey(0) == 127:      ## Skip Image 127 -> delete
        pass
    
    
   
```
